What does it mean to be an ideal gas? We created the concept of an Ideal gas as an approximation that helps us model and predict the behavior of real gases. The term ideal gas refers to a hypothetical gas composed of molecules which follow a set of key assumptions. 

1. No intermolecular interactions
2. Gas molecule volume is negligible 
3. Perfectly elastic collions

Intermolecular interactions are the forces which mediate interaction between molecules, including forces of attraction or repulsion which act between molecules and other types of neighboring particles. Basically they are the forces which hold a molecule together. There are three major types of intermolecular forces. First is the London dispersion force is also known as LDF which is the weakest of the intermolecular forces. This is the force between two nonpolar molecules. Polar bonds form when there is a difference between the electronegativity values of the atoms participating in a bond, so non-polar molecules form when atoms bond together to form molecules, they share or give electrons. If the electrons are shared equally by the atoms, then there is no resulting charge.  An example of London dispersion force is the interaction between nitrogen gas $N_2$ and oxygen gas $O_2$ molecules. The electrons of the atoms are not only attracted to their own atomic nucleus, but also to the protons in the nucleus of the other atoms. Second type is Dipole-dipole interaction, which occurs whenever two polar molecules get near each other. Dipole-dipole forces are attractive forces between the positive end of one polar molecule and the negative end of another polar molecule. The positively charged portion of one molecule is attracted to the negatively charged portion of the other. This is a common intermolecular force since many molecules are polar. An example of dipole-dipole interaction is the interaction between two sulfur dioxide $SO_2$ molecules, where the sulfur atom of one molecule is attracted to the oxygen atoms of the other molecule. Another specific example of a dipole-dipole interaction is hydrogen bonding. Hydrogen bonding is the strongest type of intermolecular interaction. It occurs when a hydrogen $H$ atom which is bound to a more electronegative atom such as nitrogen $N$, oxygen $O$, or fluorine $F$ and another adjacent atom bearing a lone pair of electrons. An example of a hydrogen bond is the dipole-dipole interaction between $H_2$ and $O$ to form $H_2O$ which is commonaly known as water. The 3rd and last general type of intermolecular interations is Ion-dipole interaction which occurs when an ion encounters a polar molecule. An ion-dipole interaction is the result of an electrostatic interaction between a charged ion and a molecule that has a dipole. It is an attractive force that is commonly found in solutions, especially ionic compounds dissolved in polar liquids. An example of ion-dipole interaction can be seen in hydrogen chloride $HCl$ where the relatively positive end  of a polar molecule ($H^+$) will attract the relatively negative end ($Cl^-$) to form a $HCl$ molecule. Van der Waals forces include attraction and repulsions between atoms, molecules, and surfaces, as well as other intermolecular forces. Van der Waals forces include intermolecular forces and also some intramolecular forces. Intramolecular forces are the forces that hold atoms together within a molecule. Intermolecular forces are forces that exist between molecules.

The second key assumption is that the volume of the molecules gas is negilible relative to the volume of the container. There are no gases that are absolutely true ideal gases because all gases molecules take up some volume. It also turns out that there will be some intermolecular or van der waal forces. Under most conditions of real gases, the first two assumptions are reasonable under relatively high temperatures and low pressures. Gas tend to become less ideal as you approach their condensation point. If you start to really low the temperature or increase the pressure you are more likely to go from a gas form to a liquid form and your gas starts to behave less and less like an ideal gas. The third assumption is that the molecules posses perfectly elastic collions. A perfectly elastic collions is defined as one in which there is no loss of kinetic energy in the collision. An inelastic collision is one in which part of the kinetic energy is changed to some other form of energy in the collision. Any collision between objects will cause it to convert some of the kinetic energy into internal energy and other forms of energy (Heat/Sound), so no large scale impacts are perfectly elastic which follows the fact that no gases are 100% ideal gases. Van der Waal adjusted the ideal gas equation $PV=nRT$ where P is the gas pressure, V is the volume of the container that the gas

Van der Waal's equation models how real gases act and their deviations from ideal gases. Real gas molecules do occupy a microscopic amount of space which differentiates them from ideal gases. Similarily real gases do have some intermolecular forces unlike ideal gases which are called van der waal forces. The devations of a real gas from an ideal gas has to do with the magnitude of the intermolecular forces and the size of the molecules. To adjust for each of these two deviations, van der waal came up with two constant (a,b) which correct for the volume of, and attractive forces between, gas molecules(Pressure). The first constant alters the pressure in the ideal gas equation. It accounts for the intermolecular attractive forces between gas molecules. The magnitude of the intermolecular forces is strongly connected to the concernation of the molecules since the more molecules we have in a set amount of volume cause more and stronger intermolecular forces(more attraction/repulation). The concentration of a substance is the quantity of solute present in a given quantity (moles per volume) of solution given by $C=\frac{n}{V}$. Each particle is attracted to one another rather than every particle being attratced only to one other particle which means the attracion forces increases exponentially with increasing concentration. Therefore we need to use $\frac{n^2}{V^2}$ to model to attraction of each particle to one another. We are going to multiply $\frac{n^2}{V^2}$ by the van der waal constant (a) in order to account for real gas intermolecular forces. We also need to account for the different characterisitcs of different gases whether it's particles attrat or repel each other. We can do that by either adding ($+a\frac{n^2}{V^2}$) or subracting ($-a\frac{n^2}{V^2}$) depending on whether the molecules are attracting or repeling each other repecitively. So the adjusted pressure is $P_{ideal}=P_{observed}\pm a\frac{n^2}{V^2}$. The other deviation between ideal and real gases is related to the size of the molecules and the free space that they take up which affects our value for the ideal volume. Each gas has particles of different sizes depending on the atoms in the gas. So each gas has an empirical volume to consider and we call this factor constant b. We need to consider the size of the molecules and we can do that by multiplying the number of moles (n) by b to get the space occupied by all of the molecules present in the gas. Therefore $V_{ideal}=V_{container}-bn$. By subtracting the space of the molecules from the volume of the container, we get the actual volume of the gas exculding the volume taken up by the particles. Unlike a, b is always postive since the molecules can't occupy negative space. After correcting the ideal gas law with these 2 constants we get Van der Waal's equation (Real gas equation):

$$(P\pm a\frac{n^2}{V^2})(V-bn)=nRT$$

The important effect that modeling particle interactions has is that this equation is able to model phase transitions. In this case, the phase transition is from a gas to a liquid. However, the transition modeled by the equation breaks
down beyond a point. To take a closer look at this behavior, let us work with the reduced gas equation. The advantage of writing the equation in reduced form is that regardless of the gas, this equation will hold, since we scaled out the constants $a$ and $b$ which are directly related to the specific gas properties. The reduced gas equation is given as

$$(P_r+\frac{3}{v^2_r})(V_r-\frac{1}{3})=\frac{8}{3}T_r $$

Solving the equation for $P_r$ as a function of $V_r$ and $T_r$ gives us

$$P_r(V_r,T_r)= \frac{8T_r}{3V_r-1}-\frac{3}{V^2_r} $$

We are going plot isotherms for the van der Waal’s equation in order to visually observe the relationship of $P_r$ and $V_r$ for a fixed temperature. An isotherm is a curve on a diagram joining points representing states or conditions of equal temperature. Plotting $P_r$ vs $V_r$ will give us the van der waal diagram which shows us how pressure and volume interact under different fixed temperatures ($T_r$ between 0.5 and 1.4).

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
t = np.linspace(0.5,1.4,10)
tr= 0.9 
v = np.linspace(0.,7,1000)
```

```python
def p(v, tr):
    return 8.0/3.0 * tr / (v- 1.0/3.0) - 3.0/(v**2)
```

```python
fig = plt.figure(figsize=(15, 7))
for tr in t:
    pr = p(v, tr)
    plt.plot(v, pr, label="$T_R = {:.3f}$".format(tr))
    
plt.ylim([-4, 5])
plt.xlabel('$v_R$')
plt.ylabel('$p_R$')
plt.legend(loc=(1.04, 0.1))
    
```

As we previously observed, when $T_R > 1$, there are no inflection points and there is only one value of $v_R$ for each value of $p_R$ as it acts as an ideal gas. Let's explore what happens as when $TR < 1$. This is where the deviations occur, for instance, consider the value $p_R$ = 0.4 when $T_R$ = 0.8. You will see that the curve passes through this value of $p_R$ at three different values of $v_R$. These 3 values are the roots of the volume which intersect a fixed value of $p_R$ on a given $T_R$ isotherm. 

We are going to convert this into a root finding problem and use the bisection method and the Newton-Raphson method to find the values of these volumes by calculating when the curve is equal to these values of pressure. 

Let's start with the bisection method

We begin with an interval that contains a root of the function. Then we cut the interval in half each time while making sure the root remains in the new interval too. 

1. Find the midpoint of the interval f(mp) between [a,b] where f(a) and f(b) have different signs
2. find f(mp) and multiply it by f(a)
3. If it is negative then take the first half as the new interval [a,mp]
4. If it is positive then take the second half as the new interval [mp,b]
5. Repeat until you get the desired amount of significant figures

Let's take an example: $f(x)=(x-2)^4-3$ in the interval [2,4]

1. mp = 3
2. f(3)= -2, f(2)= -3
3. Prodcut is positive so new interval is [3,4] with mp=3.5
4. f(3.5)=2.0625, f(3)=-2
5. Product is negative so new interval is [3,3.5] and repeat...

Now let's find the roots of the volume when $p_R = 0.4$ and when $T_R = 0.8$ using the bisection method. To find the roots we have to set our reduced function equal to 0. The reduced form can be expanded to give the following equation,

$$v_R^3 - \frac{1}{3}(1 + \frac{8T_R}{p_R})v_R^2 + \frac{3}{p_R}v_R - \frac{1}{p_R} = 0.$$

```python
def myFunction(V):
    """ Function returns the output cos(x) - x. The value of x should be in radians. 
    Keywords:
    
    x: Input. For the cosine function, this must be in radians """
    
    T = 0.8
    P = 0.4
    
    return (8*T/(3*V-1))-(3/V**2)- P
```

```python
def bisectionFunc(f=myFunction, a=0.4, b=0.6, tolerance=1e-7):
    """
    Returns the root of the input function f and the number of iterations it takes as
    a pair using the bisection method.
    
    Keywords:
    f: function for which root is to be found. Default is myFunction
    a: left most point of bracket [a, b]. Default is 0.4
    b: right most point of bracket [a, b]. Default is 1.0
    tolerance: the decimal point up to which precision is desired. Default is 1e-7 """
    
    # First figure out signs of a and b
    signa = np.sign(f(a))
    signb = np.sign(f(b))
    iter = 0
    
    # Write loop for the bisection method. Don't stop till tolerance is met
    while b-a > tolerance:
        c = (a+b)/2
        print("Best guess of root = {} at iteration {:3d}".format(c, iter))
        
        signc = np.sign(f(c))
        
        # What happens if c is zero? It means root is found!
        if signc == 0:
            return c
        elif signc == signa:
            a = c
        else:
            b = c
        iter += 1
        
        # Return final value
    return c, iter
```

```python
biroot1= bisectionFunc(myFunction, 0.4, 0.6, 1e-8)
```

```python
biroot2= bisectionFunc(myFunction, 1, 1.3, 1e-8)
```

```python
biroot3= bisectionFunc(myFunction, 3.8, 4, 1e-8)
```

By setting our intervals near where we expect our root to be, the bisection method converges to find the root to the desired amount of decimal places which is 8 in our example above

We used the following intervals to converge upon our roots 

root 1 = [0.4,0.6]

root 2 = [1,1.3]

root 3 = [3.8,4]

We can see that each root converges between those intervals from the intersection of the black line $p_R=0.4$ with our required isotherm $T_R=0.8$

Let's find the roots using Newton's method

Newton's method follows the following steps

1. The function must be differentiable (note that Bisection does not require this).
2. Using an intial approximation x[approx], calculate x[1],x[2],x[3]...
3. Using the iterative formula $$x_{i+1} = x_i - \frac{f(x_i)}{f'(x_i)}$$
4. Repeat the follwing until reached the desired number of decimal places

Let's take an example: $f(x)=(x-2)^4-3$ in the interval [2,4]

1. $$g_{x} = x - \frac{(x-2)^4-3}{4(x-2)^3}$$
2. Start with x[0]=3
3. g(3)=3.5
4. g(3.5)=3.347 and repeat...

Now let's find the roots of the volume when $p_R = 0.4$ and when $T_R = 0.8$ using Newton's method. To find the roots we have to set our reduced function equal to 0. The reduced form can be expanded to give the following equation,

$$\frac{8T_R}{3v_R-1}-\frac{3}{v_R^2}-p_R=0$$

and it's derivative is as follows 
 $$\frac{6}{v_R^3}-\frac{24T_R}{(3v_R-1)^2}=0$$

```python
def myFunction2(x=0.1):
    """
    Function returns the output cos(x) - x and its derivative as a pair of values. 
    The value of x should be in radians. 
    
    Keywords:
    x: Input. For the cosine function, this must be in radians. Default x = 0.1
    """
    T = 0.8
    P = 0.4
    
    # Return function value and derivative as a pair
    
    return (8*T/(3*x-1))-(3/x**2)- P, (6/x**3) - (24*T/((3*x-1)**2))
```

```python
def newtonMethod(fx = myFunction2, xInit = 0.5, tolerance = 1e-8):
    """
    Returns the root of the input function f and the number of iterations it takes as
    a pair using the Newton's method.
    
    Keywords:
    fx: function for which root is to be found. Default is myFunction2. fx must
       return both function value and derivative at x.
    a: left most point of bracket [a, b]. Default is 0.4
    b: right most point of bracket [a, b]. Default is 1.0
    tolerance: the decimal point up to which precision is desired. Default is 1e-7
    """
    # Initialize the variables
    
    step = 0
    f, df = fx(xInit)
    dx = -f/df
    x = xInit
    
    # Run the iterations in the algorithm
    
    while np.abs(dx) > tolerance:
        print("Best guess of root = {} at iteration {:3d}".format(x, step))
        f, df = fx(x)
        dx = -f/df
        x += dx
        step += 1
        
        # Return output
        
    return x, step
```

```python
nroot1= newtonMethod(myFunction2,0.5,1e-8 )
```

```python
nroot2= newtonMethod(myFunction2,1,1e-8 )
```

```python
nroot3= newtonMethod(myFunction2,3,1e-8 )
```

## Part 2: Complex numbers and fractals in the complex plane

In mathematics, the complex plane or z-plane is a imaginary geometric representation of the complex numbers established by the real axis and the perpendicular imaginary axis. 

![10](https://www.mathsisfun.com/algebra/images/complex-plane.svg)
* https://www.mathsisfun.com/algebra/images/complex-plane.svg

A real number, in mathematics, is a quantity that can be expressed as an infinite decimal expansion. The real numbers include the positive and negative integers and fractions (or rational numbers) and also the irrational numbers.
A complex number is a number that can be expressed in the form $a + bi$, where $a$ and $b$ are real numbers, and $i$ is a solution of the equation $i^2=1$. Complex numbers can be thought of as natural extensions to real numbers. They
also allow us to find solutions to equations that have no real number solutions. A complex number can be visually represented as a pair of numbers (a, b) forming a vector on a diagram called an Argand diagram as seen below

![11](https://upload.wikimedia.org/wikipedia/commons/thumb/5/50/A_plus_bi.svg/220px-A_plus_bi.svg.png)
* https://upload.wikimedia.org/wikipedia/commons/thumb/5/50/A_plus_bi.svg/220px-A_plus_bi.svg.png

To graph complex numbers, you simply combine the real number part and the complex part and plot them on the complex plane. In other words, given a complex number $A+Bi$, you take the real portion of the complex number $A$ to represent the x-coordinate, and you take the imaginary portion $B$ to represent the y-coordinate where the x-axis represents real numbers and the y-axis represents imaginary numbers. In the Argand coordinate plane, pure real numbers in the form $A + 0i$ exists completely on the x-axis, and pure imaginary numbers in the form $0 + Bi$ exist completely on the y-axis. The real number plane exists in the complex plane in the form of purely real numbers which lay on the x-axis, but the complex plane does not exist in the real since we have no imaginary axis in the real plane but we have a real axis in the imaginary plane. 

Since we discussed root finding techniques and introduced complex numbers, let's try to find all of the real and complex roots of the following equation

$$z^3=1$$

We know that to find the roots of any equation we have to set it equal to zero

$$z^3-1=0$$

Let's take a look on the RHS of the equation. Since we know that all real numbers exist in the imaginary plane, we can write the RHS in terms of complex numbers.

$$1=1+0i$$

Euler's formula connects complex analysis that establishes the fundamental relationship between the trigonometric functions and the complex exponential function. Euler's formula states that for any real number x:

$$e^{ix}=cos(x)+isin(x)$$

By equating our equation with Euler's formula we get

$$1=e^{0i}$$

Since this complex formula would be plotted on the complex plane, we have to think about it in terms of it's graph. We would plot $1+0i$ which just be the x-axis portion between the origin and 1. Now let's think about it's angle. Since it lies on the x-axis, it's angle is 0 or $2\pi$ or $4\pi$...

Therefore,

$$e^{0i}=e^{2\pi i}=e^{4\pi i}=1$$

$$z^3=e^{0i}=e^{2\pi i}=e^{4\pi i}=1$$

Taking the cube root of each term gives us

$$z=e^\frac{2\pi i}{3}=e^\frac{4\pi i}{3}=1$$

Visualizing these numbers will give you three vectors of magnitude 1 at 0rad$=0^\circ$,$\frac{2\pi}{3}$rad$=120^\circ$, and $\frac{4\pi}{3}$rad$=240^\circ$.
Each of these terms represents a root that can be found in the form of $A+iB$ by using Euler's formula

$$z_1=1$$

$$z_2=e^\frac{2\pi i}{3}=cos(\frac{2\pi}{3})+isin(\frac{2\pi}{3}) $$

$$z_3=e^\frac{4\pi i}{3}=cos(\frac{4\pi}{3})+isin(\frac{4\pi}{3}) $$

By simplifying the roots we get 

$$z_1=1+0i$$

$$z_2=-\frac{1}{2}+\frac{\sqrt{3}}{2}i$$

$$z_3=-\frac{1}{2}-\frac{\sqrt{3}}{2}i$$

Where $z_1=1$ is a real root and $z_2=-\frac{1}{2}+\frac{\sqrt{3}}{2}i$ and $z_3=-\frac{1}{2}-\frac{\sqrt{3}}{2}i$ are complex roots.

Now let's see if we can find these roots using the Newton method as we did in part 1

*source 11,12

```python
def myFunction3(z=0.1):
    
    return z**3-1,3*z**2
```

```python
def newtonMethod2(fx = myFunction3, xInit = 0.5, tolerance = 1e-8):
    """
    Returns the root of the input function f and the number of iterations it takes as
    a pair using the Newton's method.
    
    Keywords:
    fx: function for which root is to be found. Default is myFunction2. fx must
       return both function value and derivative at x.
    a: left most point of bracket [a, b]. Default is 0.4
    b: right most point of bracket [a, b]. Default is 1.0
    tolerance: the decimal point up to which precision is desired. Default is 1e-7
    """
    # Initialize the variables
    
    step = 0
    f, df = fx(xInit)
    dx = -f/df
    x = xInit
    
    # Run the iterations in the algorithm
    
    while np.abs(dx) > tolerance:
        print("Best guess of root = {} at iteration {:3d}".format(x, step))
        f, df = fx(x)
        dx = -f/df
        x += dx
        step += 1
        
        # Return output
        
    return x, step
```

```python
a = complex(1,0)
print(a)
```

```python
niroot1= newtonMethod(myFunction3,a,1e-8 )
print(niroot1)
```

```python
b = complex(-0.5,0.85)
print(b)
```

```python
niroot2= newtonMethod(myFunction3,b,1e-8 )
```

```python
c = complex(-0.5,-0.85)
print(c)
```

```python
niroot3= newtonMethod(myFunction3,c,1e-8 )
```

```python
roots= np.array([niroot1[0],niroot2[0],niroot3[0]])
print(roots)
```

We can verify that our roots we calculated are correct since they match the roots that the newton method gives us

Typically, gases are described by using properties like pressure, volume, and temperature. The ideal gas equation only  works for certain pressures, temperatures and volumes. It usually works in a low pressure and high temperature enviroment, but it ignores the molecular size and inter-molecular interactions which are essential for many characteristics that real gases show. Van der Waal was a phenomenal Dutch theoretical physicist and thermodynamicist who came up with an equation that shows that real gases do not follow the ideal gas law, instead they show certain deviations at their transition phases. Van der Waal's equation allows us to model real gases using a corrected ideal gas law equaiton. Ideal gases are hypothetical gases and don't fully exist in nature, therefore it is important for us the be able to model the way that actual gases act under different real life circumstances. The ideal gas equation is given by: 

$$PV=nRT $$

Van der Waal's equation is given by: 

$$(P\pm a\frac{n^2}{V^2})(V-bn)=nRT$$

Plotting $P_r$ vs $V_r$ gave us the van der waal diagram which shows us how pressure and volume interact under different fixed temperatures. We observed many interesting properties of van der waals diagram. The critical point: A point is a point on a curve where the gradient is zero. In our case it is a point on a phase diagram at which both the liquid and gas phases of a substance have the same density, and are therefore indistinguishable. The triple point: It is where the temperature and pressure at which the solid, liquid, and vapor phases of a pure substance can coexist in equilibrium. In other words, it is the point in which a substance can be solid, liquid and gas all at the same time and it is when a liquid can boil and freeze all at once.
