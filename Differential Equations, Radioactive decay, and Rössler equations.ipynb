### Radioactive decay

Radioactive decay is the process by which an unstable atomic nucleus loses energy (in terms of mass) by emitting 
radiation in the form of an alpha particle, beta particle, or gamma ray to become a stable nucleus. A material containing such unstable nuclei is considered radioactive. Certain highly excited short-lived nuclear states can decay through neutron emission, or more rarely, proton emission. Early researchers found that an electric or magnetic field could split radioactive emissions into three types of beams. The rays were given the names alpha, beta, and gamma, in increasing order of their ability to penetrate matter. Alpha decay occurs when the nucleus ejects an alpha particle. This is the most common process of emitting nucleons, but highly excited nuclei can eject single nucleons, or in the case of cluster decay, specific light nuclei of other elements.  Alpha decay is observed only in heavier elements of atomic number 52 (tellurium) and greater, with the exception of beryllium-8 which decays to two alpha particles. Beta decay occurs in two ways: (1) beta-minus decay, when the nucleus emits an electron and an antineutrino in a process that changes a neutron to a proton, or (2) beta-plus decay, when the nucleus emits a positron and a neutrino in a process that changes a proton to a neutron. Highly excited neutron-rich nuclei occasionally lose energy by way of neutron emission, resulting in a change from one isotope to another of the same element. The nucleus may capture an orbiting electron, causing a proton to convert into a neutron in a process called electron capture. All of these processes result in a well-defined nuclear transmutation. By contrast, the gamma ray decay processes does not result in a nuclear transmutation. The energy of an excited nucleus may be emitted as gamma ray in gamma ray decay; Energy may be lost when the nucleus interacts with an orbital electron causing its ejection from the atom, in a process called internal conversion. Beta and gamma rays of decay are produced by all of the elements. Radioactive decay is seen in all isotopes of all elements of atomic number 83 (bismuth) or greater.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def radioactive(N=0.1, lamb = 0.1):
    """
    Function returns the output -lamb*N. 
    
    Keywords:
    x: Input. This is the population of the species. Can be concentration too so it is ok to input floats.
    lamb: Input. This is the decay constant. Note that lambda in python has a special meaning. So we name
          this variable lamb.
    output: returns -lamb*N.
    
    """
    
    return -lamb*N
```

Now let us write down a function that performs the Euler method.

```python
def euler(func, h=0.1, yCurrent=0.1, *funcArgs):
    """
    Function performs one Euler step returns the value for the next iteration.
    
    Keywords:
    func: right hand side of the differential equation. 
          Default is the radioactivity function i.e. radioactive().

    h: Input. This is the time step used to the advance the solution to the next time point.
    yCurrent: Value of variable at current time step.
    *funcArgs: Parameters for the function in the differential equation.
    yNext: Output. Value of the variable at the next time step.
    
    """
    
    # Initialize
    yNext = 0.0
    
    # Calculate next time step
    yNext = h*func(yCurrent, *funcArgs) + yCurrent
    
    return yNext
```

We can visualize the results in the form of the graph, we used $N_0=50$, $h=0.5$, $\lambda=1.0$

```python
numIter = 50
y = np.zeros((numIter,1))
y[0] = 50.0
h = 0.5
lamb = 1.0
tList = np.linspace(0, h*numIter, numIter)

for i in range(1, len(y)):
    y[i] = euler(radioactive, h, y[i-1], lamb)
```

```python
fig, axes = plt.subplots(figsize=(12, 5))

axes.plot(tList, y,  '.', markersize=12)
axes.set_xlabel('time', fontsize=15)
plt.xlim(-1,8)
axes.set_ylabel('y', fontsize=15)
axes.grid(True)
```

After another $t_{1/2}$ the population becomes 12.5 and so on until it completely decays in the form of a geometric progression. We need to compare this with the exact solution to make sure that our approximation represents the solution. The exact solution of the equation is given by

```python
fig, axes = plt.subplots(figsize=(12, 5))

axes.plot(tList, y,  '.', markersize=12, label='Numerical solution using Euler method')
axes.plot(tList, y[0]*np.exp(-lamb*tList), '-', label='Exact solution')
axes.set_xlabel('time', fontsize=15)
axes.set_ylabel('y', fontsize=15)
plt.xlim(-1,8)
axes.grid(True)
plt.legend(fontsize=15)
```

The Runge-Kutta methods are a class of solvers studied and developed in the early 1900s by the German mathematicians C. Runge and M. W. Kutta. They build upon the discretization of the differential equations to make  and extend it to higher orders in $h$ i.e. for the same value of $h$, they are much more precise. The algorithm consists of iterating over the following steps. If you have a step size $h$, then

$$y_{n+1} = y_n + \frac{h}{6}(k_1 + 2k_2 + 2k_3 + k_4),$$

where,

$$k_1 = f(t_n, y_n),$$

$$k_2 = f(t_n + \frac{h}{2}, y_n + \frac{hk_1}{2}),$$

$$k_3 = f(t_n + \frac{h}{2}, y_n + \frac{hk_2}{2}),$$

$$k_4 = f(t_n + \frac{h}{2}, y_n + hk_3).$$

 
 
 What we are trying to do is to start at the point t0 and go to t0+h. We could do something very naive because we know the first dervivative of our equation, which is given by $k_1$ and we could take one step into the graph with that value, which would make us end up with a staright line. Instead we take half a step until we reach the midpoint $t_0+h/2$ and we calculate the dervivative at the points $d3$ or $d1$. That means that we could have taken the dervivative at the both of these points to better approximate the exact value of the coordinate at the points. We calculate both the dervivatives and find their find their midpoints which would give us the values of $k_2$ and $k_3$. To correct for the use of $k_1$ we average it with $k_2$ and the vice versa, therefore our error is minimized because it cancels itself out. We take another half step since both points we found are not enough; using both these points that are not very accurate individually, but are accurate when you take their midpoints. Now extending the step to $t_0+h$ we take the dervivative of another endpoint of the interval that forms around the desired midpoint. Now we have take 4 different points and the we know that the midpoints of each of the 2 intervals represent the accurate approximation of the solution curve. We are going to the take average of the 4 points and this will give us the approximation which is the curve between all 4 points. The RK4 method is accurate because you are continously taking the average of multiple points in order the compress the approximated curve as much as possible into the solution curve. The RK4 method is fast because we do not need to take a small step size, which requires more points on the graph, in order to get an accurate approximation because we are using an interval which converges into the accurate path of the curve rather than using many points close to each other.  The Euler method is actually the simplest version of the RK method. The most popular RK method is RK4 since it offers a good balance between order of accuracy and cost of computation. RK4 is the highest order explicit Runge-Kutta method that requires the same number of steps as the order of accuracy (RK1=1 stage, RK2=2 stages, RK3=3 stages,RK4=4 stages, ...). Beyond fourth order the RK methods become relatively more expensive to compute.

```python
 def rk4(func, h=0.1, yCurrent=0.1, *funcArgs):
    """
    Function performs one RK4 step returns the value for the next iteration.
    
    Keywords:
    func: right hand side of the differential equation. 
          Default is the radioactivity function i.e. radioactive().
    h: Input. This is the time step used to the advance the solution to the next time point.
    yCurrent: Value of variable at current time step.
    *funcArgs: Parameters for the function in the differential equation.
        
    yNext: Output. Value of the variable at the next time step.
    
    """
    
    # Initialize
    yNext = 0.0
    
    # Calculate k1, k2, k3 and k4
    
    k1 = func(yCurrent, *funcArgs)
    k2 = func(yCurrent + h*k1/2, *funcArgs)
    k3 = func(yCurrent + h*k2/2, *funcArgs)
    k4 = func(yCurrent + h*k3, *funcArgs)
    
    # Calculate next time step
    yNext = (h/6)*(k1 + 2*k2 + 2*k3 + k4) + yCurrent
    
    return yNext
```

We can visualize the results in the form of the graph, we used the same variables as we did in Eulers $N_0=50$, $h=0.5$, $\lambda=1.0$ in order to get compare them accurately

```python
numIter = 50
yRK4 = np.zeros((numIter,1))
yRK4[0] = 50.0
h = 0.5
lamb = 1.0
tList = np.linspace(0, h*numIter, numIter)

for i in range(1, len(yRK4)):
    yRK4[i] = rk4(radioactive, h, yRK4[i-1], lamb)
```

```python
fig, axes = plt.subplots(figsize=(12, 5))
axes.plot(tList, yRK4, '.',  markersize=12, label='Numerical solution using RK4 method')
axes.plot(tList, y[0]*np.exp(-lamb*tList), '-', label='Exact solution')
axes.set_xlabel('time', fontsize=15)
axes.set_ylabel('y', fontsize=15)
plt.xlim(-1,8)
axes.grid(True)
plt.legend(fontsize=15)
```

You can get the fixed points of the Rossler equations can be obtained by setting $\frac{dx}{dt}=\frac{dy}{dt}=\frac{dz}{dt}=0$. This gives the following fixed points $(x_0,y_0,z_0)$

$$x_0 = \frac{c\pm\sqrt{c^2 - 4ab}}{2}$$

$$y_0 = -\frac{c\pm\sqrt{c^2 - 4ab}}{2a}$$

$$z_0 = \frac{c\pm\sqrt{c^2 - 4ab}}{2a}$$

There are three parameters to vary here. So there is a wide variety of behavior to explore. Let's begin by setting a = b = 0.2 then trying out different values of c.

```python
a = 0.2
b = 0.2
c = 3.5

def f(state, t):
    x, y, z = state  # unpack the state vector
    return -y - z, x + a * y, b + z * (x - c)  # derivatives

state0 = [1.0, 1.0,1.0]
t = np.arange(100, 500.0, 0.01)

states = odeint(f, state0,t)

fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(states[:,0], states[:,1])
```
